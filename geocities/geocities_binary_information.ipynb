# Environment

Next, we'll setup our environment so we can load our derivatives into [pandas](https://pandas.pydata.org), build charts with [Altair](https://altair-viz.github.io/), and use the [Data Table extension for Colab](https://colab.research.google.com/notebooks/data_table.ipynb).

```python
import pandas as pd
import altair as alt
```

```python
%load_ext google.colab.data_table
```

## Initialize datasets

```python
audio = pd.read_csv("data/geocities-audio-information.csv")
audio
```

```python
pdf = pd.read_csv("data/geocities-pdf-information.csv")
pdf
```

```python
powerpoint = pd.read_csv("data/geocities-presentation-program-information.csv")
powerpoint
```

```python
spreadsheet = pd.read_csv("data/geocities-spreadsheet-information.csv")
spreadsheet
```

```python
videos = pd.read_csv("data/geocities-video-information.csv")
videos
```

```python
word = pd.read_csv("data/geocities-word-processor-information.csv")
word
```

```python
top_audio_files = audio["filename"].value_counts().head(10)
top_audio_files
```

```python
top_audio_chart = top_audio_files.plot.bar(figsize=(25, 10))

top_audio_chart.set_title("Top Audio Filenames", fontsize=22)
top_audio_chart.set_xlabel("Filename", fontsize=20)
top_audio_chart.set_ylabel("Count", fontsize=20)
```

```python
top_pdf_files = pdf["filename"].value_counts().head(10)
top_pdf_files
```

```python
top_pdf_chart = top_pdf_files.plot.bar(figsize=(25, 10))

top_pdf_chart.set_title("Top PDF Filenames", fontsize=22)
top_pdf_chart.set_xlabel("Filename", fontsize=20)
top_pdf_chart.set_ylabel("Count", fontsize=20)
```

```python
top_powerpoint_files = powerpoint["filename"].value_counts().head(10)
top_powerpoint_files
```

```python
top_powerpoint_files_chart = top_powerpoint_files.plot.bar(figsize=(25, 10))

top_powerpoint_files_chart.set_title("Top Powerpoint Files Filenames", fontsize=22)
top_powerpoint_files_chart.set_xlabel("Filename", fontsize=20)
top_powerpoint_files_chart.set_ylabel("Count", fontsize=20)
```

```python
top_spreadsheet_files = spreadsheet["filename"].value_counts().head(10)
top_spreadsheet_files
```

```python
top_spreadsheets_chart = top_spreadsheet_files.plot.bar(figsize=(25, 10))

top_spreadsheets_chart.set_title("Top Spreadsheets Filenames", fontsize=22)
top_spreadsheets_chart.set_xlabel("Filename", fontsize=20)
top_spreadsheets_chart.set_ylabel("Count", fontsize=20)
```

```python
top_video_files = videos["filename"].value_counts().head(10)
top_video_files
```

```python
top_videos_chart = top_video_files.plot.bar(figsize=(25, 10))

top_videos_chart.set_title("Top Video Filenames", fontsize=22)
top_videos_chart.set_xlabel("Filename", fontsize=20)
top_videos_chart.set_ylabel("Count", fontsize=20)
```

```python
top_word_files = word["filename"].value_counts().head(10)
top_word_files
```

```python
top_word_files_chart = top_word_files.plot.bar(figsize=(25, 10))

top_word_files_chart.set_title("Top Word File Filenames", fontsize=22)
top_word_files_chart.set_xlabel("Filename", fontsize=20)
top_word_files_chart.set_ylabel("Count", fontsize=20)
```
